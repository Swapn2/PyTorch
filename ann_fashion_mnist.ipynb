```python
import pandas as pd
from sklearn.model_selection import train_test_split
import torch
from torch.utils.data import Dataset , DataLoader
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
```

```python
torch.manual_seed(42)
```

```python
df = pd.read_csv('/content/fmnist_small.csv')
df.head()
```

```python
x = df.iloc[:,1:].values
y = df.iloc[:,0].values
```

```python
x
```

```python
x_train,x_test,y_train , y_test = train_test_split(x,y,test_size=.2,random_state =42)
```

```python
x_train = x_train/255
x_test = x_test/255
```

```python
class CustomDataset(Dataset):
  def __init__(self,features, labels):
    self.features = torch.tensor(features ,dtype=torch.float32)
    self.labels = torch.tensor(labels , dtype = torch.long)

  def __len__(self):
    return len(self.features)
  def __getitem__(self,index):
    return self.features[index] , self.labels[index]
```

```python
train_dataset = CustomDataset(x_train , y_train)
test_dataset = CustomDataset(x_test, y_test)
```

```python
train_dataset[0]
```

```python
train_loader = DataLoader(train_dataset , batch_size = 32 , shuffle=True)
test_loader = DataLoader(test_dataset , batch_size = 32 , shuffle=False)
```

```python
class MyNN(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    self.model = nn.Sequential(
        nn.Linear(num_features, 128 ),
        nn.ReLU(),
        nn.Linear(128, 64),
        nn.ReLU(),
        nn.Linear(64,10)   # by default softmax is implemented here no need to add
    )
  def forward(self, x):
    return self.model(x)
```

```python
epochs = 100
learning_rate = .1
```

```python
model = MyNN(x_train.shape[1])

loss = nn.CrossEntropyLoss()

optimizer = optim.SGD(model.parameters(),lr = learning_rate)

```

```python
for epoch in range(epochs):

  total_epoch_loss = 0
  for batch_feature , batch_labels in train_loader:
    # forward pass
    out = model(batch_feature)

    # loss
    batch_loss = loss(out,batch_labels)

    # back pass
    optimizer.zero_grad()
    batch_loss.backward()

    #
    optimizer.step()

    total_epoch_loss = total_epoch_loss + batch_loss.item()
  avg_loss= total_epoch_loss/len(train_loader)
  print(f'Epoch : {epoch +1}, Loss: {avg_loss}')
```

```python
model.eval()
```

```python
total = 0
correct = 0
with torch.no_grad():
  for batch_feature , batch_label in test_loader:
    out = model(batch_feature)
    _,predict = torch.max(out ,1)
    total = total + batch_label.shape[0]
    correct = correct + (predict == batch_label).sum().item()
print(correct/total)
```
